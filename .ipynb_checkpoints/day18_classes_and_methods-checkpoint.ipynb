```python
from IPython.core.magic import register_cell_magic

@register_cell_magic
def add_method_to(line, cell):
    cls = eval(line.strip())
    namespace = {}
    exec(cell, globals(), namespace)
    for name, obj in namespace.items():
        setattr(cls, name, obj)
```

```python
def print_time(time):
    s = f'{time.hour:02d}:{time.minute:02d}:{time.second:02d}'
    print(s)
```

```python
class Time:
    """Represent the time, of a day"""

    def print_time(self):
        s = f"{self.hour:02d}:{self.minute:02d}:{self.second:02d}"
        print(s)
```

```python
def make_time(hour, minute, second):
    time = Time()
    time.hour = hour
    time.minute = minute
    time.second = second
    return time
```

```python
start = make_time(9,45,0)
```

```python
start.print_time()
```

```python
Time.print_time(start)
```

```python
%%add_method_to Time

def time_to_int(self):
    minutes = self.hour * 60 + self.minute
    seconds = minutes * 60 + self.second
    return seconds
```

```python
Time.time_to_int(start)
```

```python
def int_to_time(seconds):
    minute, second = divmod(seconds, 60)
    hour, minute = divmod(minute, 60)
    return make_time(hour, minute, second)
```

```python
%%add_method_to Time

def int_to_time(seconds):
    minute, second = divmod(seconds, 60)
    hour, minute = divmod(minute, 60)
    return make_time(hour, minute, second)
```

```python
start = Time.int_to_time(35100)
```

```python
start.print_time()
```

```python
def add_time(time, hours, minutes, seconds):
    duration = make_time(hours, minutes, seconds)
    seconds = time_to_int(time) + time_to_int(duration)
    return int_to_time(seconds)
```

```python
%%add_method_to Time

def add_time(self, hours, minutes, seconds):
    duration = make_time(hours, minutes, seconds)
    seconds = time_to_int(seconds) + time_to_int(self)
    return Time.int_to_time(seconds)
```

```python
end = start.add_time(1,32,0)
print_time(end)
```
