```python
from jupyturtle import make_turtle,forward,left,penup,pendown,right
import math
```

```python
def square(x):
    for i in range(4):
        forward(x)
        left(90)
```

```python
make_turtle(delay = 0.02)
penup()
left(90)
forward(50)
left(150)
forward(50)
pendown()
left(30)
square(50)
penup()
left(45)
forward((50 * math.sqrt(2))/2)
left(45)
forward(5)
left(45)
forward(5)
left(45)
pendown()
square(10)
left(180)
penup()
forward(29)
left(180)
pendown()
forward(16)
left(90)
forward(7)
left(90)
forward(16)
left(90)
penup()
forward(24)
left(90)
forward(50)
pendown()
left(45)
forward((50 * math.sqrt(2))/2)
left(90)
forward((50 * math.sqrt(2))/2)
penup()
right(90)
forward(40)
pendown()
square(20)
```
