```python
# slow method
word_list = open('words.txt').read().split()
len(word_list)
```

```python
def reverse_word(word):
    return ''.join(reversed(word))
```

```python
def count_word():
    count = 0
    for word in word_list:
        if reverse_word(word) in word_list:
            count += 1
    return count
```

```python
count_word()
```

```python
len(word_list)**2
```

```python
# making this function more faster by using dictionary
word_dict = {}
for word in word_list:
    word_dict[word] = 1
```

```python
def faster():
    count = 0
    for word in word_dict:
        if reverse_word(word) in word_dict:
            count += 1
    return count
```

```python
faster()
```

```python
def count_values(string):
    count = {}
    for letter in string:
        if letter not in count:
            count[letter] = 1
        else:
            count[letter] += 1
    return count
```

```python
string = count_values('bbssdfkjogkrr')
print(string)
```

```python
# fibonacci series
def fibonacci(n):
    if n == 0:
        return 0
    
    if n == 1:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(3)
```

```python
dict = {0:0,1:1}
def fibonacci_memo(n):
    """memo version of fibonacci series."""
    if n in dict:
        return dict[n]
    
    compute = fibonacci_memo(n-1) + fibonacci_memo(n-2)
    dict[n] = compute
    return compute
```

```python
fibonacci_memo(40)
```

```python
def count_values(string):
    count = {}
    for letter in string:
        count[letter] = count.get(letter,0) +  1
    return count
```

```python
count_values('bsddfkeeee')
```

```python
def duplicates(strings):
    dic = {}
    for string in strings:
        if string in dic:
            return True
        dic[string] = 1
    return False
```

```python
print(duplicates("unpredictably"))  
print(duplicates("dermatoglyphics")) 
print(duplicates("programming"))     
print(duplicates([1, 2, 3, 4, 1]))    
```

```python
def find_repeats(counter):
    repeats = []
    for key in counter:
        if counter[key] > 1:
            repeats.append(key)
    return repeats
```

```python
counter = {
    'b': 1,
    'r': 2,
    'o': 2,
    'n': 1,
    't': 1,
    's': 2,
    'a': 1,
    'u': 2
}

print(find_repeats(counter))
```

```python
counter1 = count_values('brontosaurus')
counter2 = count_values('apatosaurus')
```

```python
def add_counters(counter1, counter2):
    new_dict = {}
    for key in counter1:
        new_dict[key] = new_dict.get(key,0) + counter1[key]

    for key in counter2:
        new_dict[key] = new_dict.get(key,0) + counter2[key]
    return new_dict
```

```python
add_counters(counter1, counter2)
```

```python
word_set = [
    "shoe", "cold", "schooled", "school", "cat", "dog"
]
```

```python
def is_interlocking(word):
    first = word[0::2]
    second = word[1::2]
    return first in word_set and second in word_set
```

```python
is_interlocking("cat")
```
