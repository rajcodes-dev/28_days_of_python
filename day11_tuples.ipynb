```python
def trimmed_mean(*args):
    low, high = min_max(args)
    trimmed = list(args)
    trimmed.remove(low)
    trimmed.remove(high)
    return mean(*trimmed)
```

```python
trimmed_mean(1,2,3,10)
```

```python
def value_counts(string):
    counter_dict = {}
    for word in string:
        if word not in counter_dict:
            counter_dict[word] = 1
        else:
            counter_dict[word] += 1
    return counter_dict
```

```python
counter = value_counts('banana')
counter
```

```python
items = counter.items()
items
```

```python
sorted(items)
```

```python
def invert_dict1(d1):
    new_dict = {}
    for key, value in d1.items():
        if value not in new_dict:
            new_dict[value] = [key]
        else:
            new_dict[value].append(key)
    return new_dict
```

```python
letters = 'abcdefghijklmnopqrstuvwxyz'
letters_map = dict(zip(letters,range(len(letters))))
```

```python
def shift_word(word, shift):
    result = []
    for ch in word:
        old_index = letters_map[ch]
        new_index = (old_index + shift) % 26
        result.append(letters[new_index])
    return ''.join(result)
```

```python
print(shift_word("cheer", 7))   # jolly
print(shift_word("melon", 16))  # cubed
```

```python
def most_frequent_letters(string):
    freq = {}
    for ch in string:
        freq[ch] = freq.get(ch, 0) + 1
    
    temp = []
    for letter, count in freq.items():
        temp.append((letter,count))

    temp.sort(reverse = True)
    
    for letter, count in temp:
        print(letter,count)
```

```python
most_frequent_letters('hello')
```

```python
def word_anagrams(l1):
    anagrams = {}
    
    for w in l1:
        key = ''.join(sorted(w))
        if key not in anagrams:
            anagrams[key] = []
        anagrams[key].append(w)

    for group in anagrams.values():
        if len(group) > 1:
            print(group)
```

```python
words = ["deltas","desalt","lasted","salted","slated","staled",
         "retainers","ternaries",
         "generating","greatening",
         "resmelts","smelters","termless"]

word_anagrams(words)
```

```python
def word_distance(word1, word2):
    count = 0
    for x, y in zip(word1, word2):
        if x != y:
            count += 1
    return count
```

```python
word_distance('apple','ample')
```

```python
def metathesis_pair(words):
    anagrams = {}
    
    for w in words:
        key = ''.join(sorted(w))
        if key not in anagrams:
            anagrams[key] = []
        anagrams[key].append(w)

    for group in anagrams.values():
        if len(group) < 2:
            continue

        for i in range(len(group)):
            for j in range(i+1, len(group)):
                if word_distance(group[i], group[j]) == 2:
                    print(group[i], group[j])
```

```python
words = ["converse", "conserve",
         "least","slate","stale","steal","tales",
         "listen","silent"]

metathesis_pair(words)
```
