```python
def time_to_int(time):
    """Convert time to seconds."""
    minutes = time.hour * 60 + time.minute
    seconds = minutes * 60 + time.second
    return seconds
```

```python
time = make_time(1,1,1)
print_time(time)
time_to_int(time)
```

```python
def int_to_time(seconds):
    """Convert seconds to time."""
    minute, second = divmod(seconds, 60)
    hour, minute = divmod(minute, 60)
    return make_time(hour, minute, second)
```

```python
time = int_to_time(3661)
print_time(time)
```

```python
def add_time(time, hours, minutes, seconds):
    """Consice version of add time."""
    duration = make_time(hours, minutes, seconds)
    seconds = time_to_int(time) + time_to_int(duration)
    return int_to_time(seconds)
```

```python
start = make_time(9, 40, 0)
end = add_time(start, 1, 32, 0)
print_time(end)
```

```python
def substract_time(t1, t2):
    """Interval between two times."""
    t1 = time_to_int(t1)
    t2 = time_to_int(t2)
    return t2 - t1
```

```python
"""
Q.2 Write a function called is_after that takes two Time objects and returns
True if the second time is later in the day than the first, and False
otherwise:
"""
```

```python
def is_after(t1, t2):
    """Checks whether `t1` is after `t2`.
    >>> is_after(make_time(3, 2, 1), make_time(3, 2, 0))
    True
    >>> is_after(make_time(3, 2, 1), make_time(3, 2, 1))
    False
    >>> is_after(make_time(11, 12, 0), make_time(9, 40, 0))
    True
    """
    if t1.hour > t2.hour:
        return True
    elif t1.hour < t2.hour:
        return False
    
    if t1.minute > t2.minute:
        return True
    elif t1.minute < t2.minute:
        return False
    
    return t1.second > t2.second
```
