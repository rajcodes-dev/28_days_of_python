```python
from IPython.core.magic import register_cell_magic

@register_cell_magic
def add_method_to(line, cell):
    cls = eval(line.strip())
    namespace = {}
    exec(cell, globals(), namespace)
    for name, obj in namespace.items():
        setattr(cls, name, obj)
```

```python
class Card:
    """Represents a standard playing cards."""
    # class variables
    suit_names = ['Clubs','Diamonds','Hearts','Spades']
    rank_names = [None, 'Ace', 2, 3, 4, 5, 6, 7, 8, 9, 10, 
                  'Jack','Queen','King','Ace']
```

```python
Card.rank_names[11]
```

```python
%%add_method_to Card

def __init__(self, suit, rank):
    self.suit = suit
    self.rank = rank
```

```python
queen = Card(1,12)
```

```python
queen.suit, queen.rank
```

```python
queen.rank_names
# but we should use class name to access class variables because 
# object access should indicate that we are calling attributes.
```

```python
%%add_method_to Card

def __str__(self):
    rank_name = Card.rank_names[self.rank]
    suit_name = Card.suit_names[self.suit]
    return f"{rank_name} of {suit_name}"
```

```python
print(queen)
```

```python
queen2 = Card(1,12)
print(queen2)
```

```python
%%add_method_to Card

def __eq__(self, other):
    return self.suit == other.suit and self.rank == other.rank
```

```python
%%add_method_to Card

def to_tuple(self):
    return self.suit, self.rank
```

```python
%%add_method_to Card

def __lt__(self, other):
    return self.to_tuple() < other.to_tuple()
```

```python
queen < queen2
```

```python
queen > queen2
```

```python
%%add_method_to Card

def __le__(self, other):
    return self.to_tuple() <= other.to_tuple()
```

```python
queen <= queen2
```

```python
queen <= six
```

```python
queen >= six
```

```python
class Deck:
    """Represent a Deck of an cards."""

    def __init__(self, cards):
        self.cards = cards
```

```python
%%add_method_to Deck

def make_cards():
    cards = []
    for suit in range(4):
        for rank in range(2,15):
            card = Card(suit, rank)
            cards.append(card)
    return cards
```

```python
cards = Deck.make_cards()
deck = Deck(cards)
len(cards)
```

```python
%%add_method_to Deck

def __str__(self):
    res = []
    for card in self.cards:
        res.append(str(card))
    return '\n'.join(res)    
```

```python
small_deck = Deck([queen, six, queen2])
```

```python
print(small_deck)
```

```python
str(small_deck)
```

```python
%%add_method_to Deck

def take_card(self):
    return self.cards.pop()
```

```python
card = deck.take_card()
print(card)
```

```python
print(len(cards))
```

```python
%%add_method_to Deck
def put_card(self, card):
    self.cards.append(card)
```

```python
deck.put_card(card)
len(deck.cards)
```

```python
import random
```

```python
%%add_method_to Deck

def shuffle(self):
    random.shuffle(self.cards)
```

```python
deck.shuffle()
for card in deck.cards[:4]:
    print(card)
```

```python
%%add_method_to Deck

def sort(self):
    self.cards.sort()
```

```python
deck.sort()
for card in deck.cards[:4]:
    print(card)
```
