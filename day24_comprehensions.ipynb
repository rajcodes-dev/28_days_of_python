```python
values = []
for x in range(6):
    values.append(x)
```

```python
values
```

```python
values = [x for x in range(6)]
```

```python
values
```

```python
values = [x + 1 for x in range(10)]
```

```python
values
```

```python
evens = []
for number in range(50):
    if number % 2 == 0:
        evens.append(number)
```

```python
evens
```

```python
evens = [number for number in range(50) if number % 2 == 0]
```

```python
evens
```

```python
options = ["any", "albany", "apple", "world", "hello", ""]
valid_strings = []

for string in options:
    if len(string) <= 1:
        continue
    if string[0] != 'a':
        continue
    if string [-1] != 'y':
        continue
    valid_strings.append(string)
```

```python
valid_strings
```

```python
valid_strings = [
    string 
    for string in options
    if len(string) >= 2
    if string[0] == 'a'
    if string[-1] == 'y'
]
```

```python
valid_strings
```

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
flattened = []

for row in matrix:
    for num in row:
        flattened.append(num)

flattened = [num for row in matrix for num in row]
print(flattened)
```
