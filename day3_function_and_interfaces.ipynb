```python
def polygon(sides, length):
    """Create a function, to draw a polygon."""
    angle = 360/sides 
    # Angle betweent adjacent sides of the polygon is 360 / no. of sides.
    for i in range(sides):
        forward(length)
        left(angle)
```

```python
make_turtle()
polygon(sides = 8,length = 30)# calling function with keyword arguments.
```

```python
# Drawing a approximate circle 
import math

def circle(radius):
    circumference = radius * 2 * math.pi
    n = 30
    length = circumference / n
    polygon(n,length)
```

```python
make_turtle(delay = 0.02)
circle(10)
```

```python
def poly_line(n, length, angle):
    """Draw a no. of line segments, the length of LS and the angle btw LS"""
    for i in range(n):# where LS is line segment.
        forward(length)
        left(angle)
```

```python
make_turtle()
poly_line(10, 3, 360)
```

```python
def polygon(n, length):
    # Draws a polygon 
    angle = 360.0 / n
    poly_line(n, length, angle)
```

```python
make_turtle()
polygon(30, 10)
```

```python
def arc(radius, angle):
    # This funtion draw an arc using two parameters radius and angle
    arc_length = 2 * math.pi * radius * (angle/360)# Formula for arc_length
    n = 30#This constant is using to make the value small and the arc smooth.
    length = arc_length / n  
    step_angle = angle / n
    poly_line(n, length, step_angle)
```

```python
def circle(radius):
    angle = 360
    arc(radius, angle)
```

```python
make_turtle(delay = 0.02)
# we had refactor all the function 
polygon(20,9)
arc(70, 70)
circle(10)
```

```python
from jupyturtle import penup, pendown
```

```python
def jump(length):
    """Move forward length units without leaving a trail.
    Postcondition: Leaves the pen down."""
    penup()
    forward(length)
    pendown()
```

```python
make_turtle()
jump(40)
```

```python
def rectangle(length, width):
    """Draws a Rectangle using parameters length and width."""
    for i in range(2):
        forward(length)
        left(90)
        forward(width)
        left(90)
```

```python
make_turtle()
rectangle(80, 40)
```

```python
"""Write a function called rhombus that draws a rhombus with a given side length and a
given interior angle. For example, here’s a rhombus with side length of 50 and an inte
rior angle of 60 degrees:"""
def rhombus(length, int_angle):
    for i in range(2):
        forward(length)
        left(int_angle)
        forward(length)
        left(180 - int_angle)
```

```python
make_turtle()
rhombus(40, 60)
```

```python
def parallelogram(length, width, angle):
    """Draws a parallelogram by length, width, and angle as a parameter."""
    for i in range(2):
        forward(length)
        left(angle) 
        forward(width)
        left(180-angle)
```

```python
def rectangle(length, width):
    """Draws a rectangle with length and width as a parameter."""
    parallelogram(length, width,90)
```

```python
def rhombus(length):
    """Draws a rhombus with length as a parameter."""
    parallelogram(length, length, 60)
```

```python
def triangle(r, angle):
    """
    Draws one triangular slice with radius r and central angle.
    """
    base = 2 * r * math.sin(math.radians(angle / 2))

    forward(r)
    left(90 + angle / 2)
    forward(base)
    left(90 + angle / 2)
    forward(r)
    left(180 - angle)
```

```python
def draw_pie(n, r):
    """
    Draws a pie made of n triangular slices.
    """
    angle = 360 / n

    for i in range(n):
        triangle(r, angle)
        left(angle)
```

```python
make_turtle()
draw_pie(8, 50)
```

```python
"""Write an appropriately general set of functions that can draw 
flowers like this.
Hint: use arc to write a function called petal that draws one flower petal."""
```

```python
def poly_line(n, radius, angle):
    for i in range(n):
        forward(radius)
        left(angle)
```

```python
def arc(radius, angle):
    n = 30
    arc_length = 2 * math.pi * radius * (angle/360)
    length = arc_length/n
    step_angle = angle/n
    poly_line(n, length, step_angle)
```

```python
def petal(radius, angle):
    arc(radius, angle)
    left(180 - angle)
    arc(radius, angle)
    left(180 - angle)
```

```python
def flower(n, radius, angle):
    for i in range(n):
        petal(radius, angle)
        left(360/n)
```

```python
make_turtle(delay = 0.002)
flower(12,80,60)
```
