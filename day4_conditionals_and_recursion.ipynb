```python
from time import time

now = time()
now 
```

```python
seconds_in_day = 24 * 3600
seconds_in_hours = 3600
seconds_in_minute = 60
```

```python
days = now // seconds_in_day
remain_seconds = now % seconds_in_day
hours = remain_seconds // seconds_in_hours
remain_minutes = remain_seconds % seconds_in_hours
minutes = remain_minutes // seconds_in_minute
seconds = remain_minutes % seconds_in_minute
```

```python
"""If you are given three sticks, you may or may not be able to arrange them in a trian
gle. For example, if one of the sticks is 12 inches long and the other two are 1 inch
long, you will not be able to get the short sticks to meet in the middle. For any three
lengths, there is a test to see if it is possible to form a triangle:
If any of the three lengths is greater than the sum of the other two, then you cannot
form a triangle. Otherwise, you can. (If the sum of two lengths equals the third, they
form what is called a “degenerate” triangle.)
Write a function named is_triangle that takes three integers as arguments, and that
prints either “Yes” or “No,” depending on whether you can or cannot form a triangle
from sticks with the given lengths. Hint: use a chained conditional."""
```

```python
def is_triangle(s1, s2, s3):
    """Tells is that the triangle can me formed or not."""
    if(s1 >= s2 + s3 or s2 >= s1 + s3 or s3 >= s1 + s2):
        print("NO!")
    else:
        print("YES!")
```

```python
is_triangle(50,20,60)
```

```python
from jupyturtle import forward, left, right, back

def draw(length):
    angle = 50
    factor = 0.6
    if length > 5:
        forward(length)
        left(angle)
        draw(factor * length)
        right(2 * angle)
        draw(factor * length)
        left(angle)
        back(length)
```

```python
"""To draw a Koch curve with length x, all you have to do is:
1. Draw a Koch curve with length x/3.
2. Turn left 60 degrees.
3. Draw a Koch curve with length x/3.
4. Turn right 120 degrees.
5. Draw a Koch curve with length x/3.
6. Turn left 60 degrees.
7. Draw a Koch curve with length x/3.
The exception is if x is less than 5—in that case, you can just draw a straight line with
length x.
Write a function called koch that takes x as a parameter and draws a Koch curve with
the given length. The result should look like this:"""
```

```python
def koch(x):
    "Draws a koch curve."
    if x < 5:
        forward(x)
    else:
        koch(x/3)
        left(60)
        koch(x/3)
        right(120)
        koch(x/3)
        left(60)
        koch(x/3)
```

```python
make_turtle(delay=0)
koch(120)
```

```python
def triangle(length):
    for _ in range(3):
        forward(length)
        left(120)
```

```python
def sierpinski(length, n):
    if n == 0:
        triangle(length)
    else:
        sierpinski(length/2, n-1)
        forward(length/2)
        sierpinski(length/2, n-1)
        back(length/2)
        left(60)
        forward(length/2)
        right(60)
        sierpinski(length/2, n-1)
        left(60)
        back(length/2)
        right(60)
```
